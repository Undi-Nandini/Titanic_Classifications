**DATA PREPROCESSING**

Imputation(Handling missing values)

```python
#droping the Cabin from data
data.drop("Cabin",axis=1,inplace=True)
```

```python
data.head()
```

```python
#droping null values from data
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

ONE-HOT ENCODING Changing categorical values into binary values

```python
sex = pd.get_dummies(data['Sex'], drop_first=True)
sex.head(5)
```

```python
pcl=pd.get_dummies(data["Pclass"],drop_first=True)
pcl.head(5)
```

```python
embark=pd.get_dummies(data["Embarked"],drop_first=True)
embark.head(5)
```

```python
#Concatination
df=pd.concat([data,sex,embark,pcl],axis=1)
df.head(5)
```

```python
data.drop(["Embarked","Pclass","PassengerId","Name","Ticket","Sex"],axis=1,inplace=True)
```

```python
data.head(5)
```

```python
X=data.drop("Survived",axis=1)
X
```

```python
Y=data["Survived"]
Y
```

**MODEL TRAINING**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.svm import SVC
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X.columns = X.columns.astype(str)
```

```python
X=X.values
```

```python
classifier = SVC()
classifier.fit(X_train,Y_train)
```

```python
 predictions =classifier.predict(X_test)
```

```python
print(classification_report(Y_test,predictions))
```

**MODEL EVALUATION**

```python
conf_matrix = confusion_matrix(Y_test, predictions)
```

```python
print(confusion_matrix(Y_test,predictions))
```

```python
# Assuming X_train is your new, unseen data
new_predictions = classifier.predict(X_train)
```

```python
sns.heatmap(conf_matrix, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()

```
